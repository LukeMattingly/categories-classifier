# B&H Category Mapping

Load the Categories for B&H

```python
categories = []

with open('b&h_categories.txt', 'r') as file:
    for line in file:
        category = line.strip()
        categories.append(category)
```

```python
categories
```

Load the list of all topics and their associated uuids

```python
id_name_mapping = {}

with open('master_topics.csv', 'r') as file:
    for line in file:
        values = line.strip().split(',')
        if len(values) >= 2:
            uuid, topic = values[0], ','.join(values[1:])
            id_name_mapping[uuid] = topic
```

```python
id_name_mapping
```

Associate the list of names from the unique ids in the b&h topics list(unique_text.txt) with master topic to get their topic names

```python
# Create a list to store the associated names
associated_names = []

# Read the unique IDs from the "unique_ids.txt" file
with open('unique_ids.txt', 'r') as file:
    unique_ids = [line.strip(',\n') for line in file]

# Loop through the unique IDs and look up their associated names
for unique_id in unique_ids:
    if unique_id in id_name_mapping:
        associated_names.append(id_name_mapping[unique_id])
    else:
        # Handle the case when the unique ID is not found in the mapping
        associated_names.append('Not Found')

print(associated_names)
```

```python
unique_ids
```

```python
candidate_labels = associated_names
```

Do classification on the b&h categories list

```python
for sequence_to_classify in categories[:10]:
    result = classifier(sequence_to_classify, candidate_labels, multi_label=True)
    print(result)
```
